```python
import torch as t
a = [[i,j] for i in range(10) for j in range (10)]
a.extend([[0,0],[0,0]])
print(a)
a = t.tensor(a)
print(a.size())
```

```python
# 现在的任务就是从a中过滤掉 重复的值对
x = a[:,0]
y = a[:,1]
mask = (x!=y) # 
print(mask) # 是否过滤
print(mask.size()) 
idx = mask.nonzero().view(-1) # 得到过滤后的下标
print(idx.size())
res = a[idx]
print(res)
print(res.size())
```
