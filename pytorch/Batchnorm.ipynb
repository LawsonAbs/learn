```python
'''
(1)BatchNorm 不会修改输入tensor的size
（2）下面这个在二维tensor上计算BatchNorm 是正确的， 但是为啥迁移到3维就不对了呢？
'''
import torch.nn as nn
import torch

# Without Learnable Parameters
m = nn.BatchNorm1d(60, affine=False)
inp = torch.randn(20,60) # (bsz, feature_dims, seq_length)
print(inp)
output = m(inp)
print(output.size())
print(output)
```

```python
eps = 1e-05
mean = torch.mean(inp,dim=0)
var = torch.var(inp,dim=0,unbiased=False)
print(mean.size())
print(var.size())
print(mean)
print(var)
res = (inp-mean)/torch.sqrt(var + eps)
print(res)
```
