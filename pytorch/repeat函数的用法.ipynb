```python
'''
使用repeat 函数
功能：按照指定维度进行相同扩充
'''
import torch 
x = torch.tensor([1, 2, 3])
print(x.size())
y = x.repeat(4,2,3)
print(y.size())
print(y)
```

```python
'''
repeat函数有什么用处呢？下面举个例子：
场景一：现在有个二维矩阵a，里面存储的是每个实体的embedding，维度是3，实体个数是4。现在需要判断这4个实体相互之间的一个关系，那么就会组成4*4个实体对。常见做法是用双重for循环，
写法如下：
'''
a = torch.randn(4,3)
print(a)
# 现在对上面这个进行一个扩充
h = [] # 头实体
t = [] # 尾实体
for i in range(4):
    for j in range(4):
        h.append(a[i])
        t.append(a[j])
print(h)
print(t)
```

```python
'''
但是有改进之处吗？显然这种简单的逻辑可以用其它函数来代替，这个函数就是repeat，且看如下操作，操作结果会发现和上面是一模一样。但是速度会有大大提升。
'''
t_mention = a.repeat(a.size(0),1) # 扩充一下
print(t_mention)
h_mention = a.repeat(1,a.size(0))
print(h_mention.view(-1,3))
```
