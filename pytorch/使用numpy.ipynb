```python
#numpy切片

import numpy as np
import torch

a = torch.rand(10,2)
print(a)

b = a.numpy()

print(b) # 输出b 
print("===============")
print(b[:,0])# 
print("===============")
print(b[8:]) # 从第8维开始输出
print("===============")
```

```python
import numpy as np
 
a = np.array([[1,2,3],[3,4,5],[4,5,6]])  
print(a)
print("======列处理======")
print (a[...,1])   # 第2列元素
print("============")
print(a[:,1]) # 第2列元素



print("======行处理======")
print (a[1,...])   # 第2行元素
print("============")
print(a[1,:])  # 第2行元素
print("============")
print (a[...,1:])  # 第2列及剩下的所有元素
```
