```python
"""
1.plot
pycharm 
"""
import matplotlib.pyplot as plt
import random
import torch
import numpy as np

num = 0  # 
#plt.ion()  # 
while num < 10:
    plt.clf()  # 
    x = np.linspace(-3, 3, 50) # 
    w = random.randint(1,8)  # 
    y = w * x + 1
    # plt.figure(num=3, figsize=(8, 5))  # figsize
    plt.plot(x, y, color='red', linewidth=2.0, linestyle='dashdot')
    plt.xlim((-1, 2))  # x
    plt.ylim((-2, 3))  # y
    plt.xlabel('x')  # x
    plt.ylabel('y')  # y
    new_ticks = np.linspace(-1, 2, 5)
    plt.xticks(new_ticks)  # x
    # y
    plt.yticks([-2, -1, 0, 1, 2, 3]),
               #[r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])

    plt.pause(0.4)  # 
    num = num + 1

plt.ioff()  # 
plt.show()  # 
```
