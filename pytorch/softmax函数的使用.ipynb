```python
import torch
import torch.nn as nn
dim0 = nn.Softmax(0)
dim1 = nn.Softmax(1)
dim2 = nn.Softmax(2)
dim_1 = nn.Softmax(-1)
dim_2 = nn.Softmax(-2)
```

```python
temp = torch.tensor([0.,4.,4.])
print(dim0(temp))
```

```python
"""
先看一维
"""
a = [1.,2.,3.]
a = torch.tensor(a)

print(dim0(a)) # 
# print(dim1(a)) # IndexError 
# print(dim2(a)) # IndexError Dimension out of range (expected to be in range of [-1, 0], but got 2)
print(dim_1(a)) 
# print(dim_2(a)) #IndexError: Dimension out of range (expected to be in range of [-1, 0], but got -2)
```

```python
"""
再看二维
"""
a=torch.randn(2,3)
print(a)
print("====================================")
print(dim0(a)) # 将(2,3)看成是2个1*3的矩阵。将这两个矩阵上的数取softmax便得到结果
print(dim1(a)) # 
#print(dim2(a)) # IndexError: Dimension out of range (expected to be in range of [-2, 1], but got 2)
print(dim_1(a))
print(dim_2(a))
```

```python
"""
再看三维
"""
a=torch.randn(3,2,2)
print(a)
print("====================================")
print(dim0(a)) # 将(3,1,2)看成是3个1*2的矩阵。将这三个矩阵上的对应位置的数取softmax便得到结果
print(dim1(a)) 
print(dim2(a)) 
print(dim_1(a))
print(dim_2(a))
```

```python
"""
再看二维
"""
a=torch.randn(2,3,3,4,5)
dim3=nn.Softmax(3)
print(a)
print("====================================")
print(dim0(a)) #可以将这个矩阵看成是2个(3,3,4,5) 合并的效果。dim0的效果就是将这两个不同的矩阵对应位置的数联合取softmax
print(dim1(a)) 
print(dim2(a)) 
print(dim3(a))
print(dim_1(a))
print(dim_2(a)) 
```

```python
row = nn.Softmax(dim=1) # 按行方式  => 即每行都处理
input = torch.randn(1,3)
output = row(input)
print(input)
print("=====")
print(output)

col = nn.Softmax(dim = 0) # 按列方式 => 即每列都处理
col_res = col(input)
print("=====")
print(col_res)

minus = nn.Softmax(-1)  # 也是按行方式
min_res = minus(input)
print("=====")
print(min_res)
```
