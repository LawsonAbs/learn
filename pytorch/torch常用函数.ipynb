```python
import torch as t
"""tensor维度的操作
"""
a = t.randn(3,4,2)
print(a)
print(a[-1]) # 取得是a[2][..][...]的数据
c = a[-1]
print(c.size())# torch.Size([4,2])
```

```python
b = a[-1][-1]
print(b)
```

```python
"""求tensor的长度，即len()函数
=》返回的是tensor第一维的长度
"""

a = t.randn(12,3,4)
print(len(a))  # 这个输出的是tensor的第一个维度的长度
```

```python
"""
1.torch.cat()函数的使用
01.doc： https://pytorch.org/docs/stable/generated/torch.cat.html?highlight=cat#torch.cat
"""
import torch

x = torch.randn(2, 3)
y = torch.randn(2,5)
print(x)
print(y)
print("---------------")
res1 = torch.cat((x, x), 0) # 在第0维拼接  => 必须保证其他维度相同
print(res1 )

res2 = torch.cat((x, y), 1) # 在第1维拼接 =>  保证其他维相同，但是拼接的那一维可以不用相同
print(res2)
 
```

```python
"""
item()函数
"""
import torch
a = torch.randn(1)
print(a)
print(a.data)
b = a.item()  # item() 函数将tensor转成一个数字 
print(b)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch

# 假数据
n_data = torch.ones(100, 2)         # 数据的基本形态
x0 = torch.normal(2*n_data, 1)      
x1 = torch.normal(-2*n_data, 1)     

y0 = torch.zeros(100)               
y1 = torch.ones(100)                
 
# 注意 x, y 数据的数据形式是一定要像下面一样 (torch.cat 是在合并数据)
x = torch.cat((x0, x1), 0).type(torch.FloatTensor)  # FloatTensor = 32-bit floating
y = torch.cat((y0, y1), 0).type(torch.FloatTensor)    # LongTensor = 64-bit integer
 
# 根据（横坐标，纵坐标）画图
plt.scatter(x.data.numpy()[:, 0], 
            x.data.numpy()[:, 1],
            c=y.data.numpy(), 
            s=100, lw=0, cmap='RdYlGn')
plt.show()
```
