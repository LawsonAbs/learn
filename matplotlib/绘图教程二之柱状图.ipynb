```python
"""
单副图片
"""
import numpy as np
import os
import matplotlib.pyplot as plt

x = ['2-6', '1-4', '1-9', '1-1', '5-30', '2-33', '2-3', '6-15', '6-34', '1-10', '2-2', '6-8' ] 
y = [8,2,10,4,5,1,7,18,9,10,11,2]
# print(x)
index = np.arange(len(x))
plt.figure(figsize=(30,10))

# 在图像上写文本
for a,b in zip(x,y):    
    plt.text(a, b+0.05, '%.0f' % b, ha='center', va='bottom',fontsize=17)

# 调节x轴刻度字体大小
plt.tick_params(labelsize=15)
plt.xlabel("class",fontdict={"fontsize":17})
plt.ylabel("num",fontdict={"fontsize":17})
plt.bar(x,y,width=0.3,align='center') # 画柱状图
plt.show()
```
