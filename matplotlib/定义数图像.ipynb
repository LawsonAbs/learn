```python
import numpy as np
from pylab import * # 调整子图间距
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter # 修改刻度
from matplotlib import rc

#创建变量数组，用于画图
x1= np.array([0,5,10,15,20,25,30,35,40,45,50])
y1= np.array([0.8994,0.8995,0.8925,0.9025,0.9042,0.9003,0.9017,0.9039,0.9025,0.9047,0.9048])
y2= np.array([0.9228,0.9244,0.9237,0.9181,0.9265,0.9275,0.9300,0.9259,0.9272,0.9285,0.9285])
#创建图形
plt.figure(figsize=(6.4,4.8))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
    

#在index=1处（第一行第一列）画图
ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)

xmajorLocator = MultipleLocator(10) # 将x主刻度标签设置为10的倍数
xminorLocator = MultipleLocator(5) # 将x主刻度标签设置为10的倍数

# step2.设置画什么图
plt.sca(ax1)
plt.plot(x1,y1,'--^',color='red')
plt.legend(["homographic"],loc=4)
plt.xlabel(r'$d_S$')
plt.ylabel(r'F1')
ax1.xaxis.set_major_locator(xmajorLocator)
ax1.xaxis.set_minor_locator(xminorLocator)


plt.sca(ax2)
plt.plot(x1,y2,'-*',color='blue') #
plt.legend(["heterographic"],loc=4)
plt.xlabel(r'$d_S$')
plt.ylabel('F1')
ax2.xaxis.set_major_locator(xmajorLocator)
ax2.xaxis.set_minor_locator(xminorLocator)
ax1.set_ylim(0.89,0.91)
ax2.set_ylim(0.91,0.935)

# 调整两个子图的间距
subplots_adjust(hspace=0.1,wspace=0.45)
name='./defi_num.png'
# 300参数用于保真
plt.savefig(name,dpi=300) # 存储图片
plt.show()
```
