```python
'''使用matplotlib 画一个简单的函数图像
01.一个方框只有一条曲线
02.使用x,y两个list画出一条曲线，其中y的值既可以是已知的，也可以是对x按照某种函数操作生成
'''
import matplotlib.pyplot as plt
from math import log
x= [1,2,3,4,5]
# print(x)
y = [log(i) for i in x]
# print(y)
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
macro_f1 =  [0.42556946852969163,0.4350632492487473,0.4281398421414817,0.4299698034777477,  0.4225653800933473,  0.4201582555375919,  0.3774468897079784,  0.3695882112560943,  0.29766456731542057,  0.04822709519865576]
macro_f1.reverse() # 将list 倒序，这里是对list原地修改
x = [i for i in range(len(macro_f1))]
# print(x)
plt.plot(x,macro_f1)
plt.show()
```

```python
'''
一个方框中有多条曲线
'''
import matplotlib.pyplot as plt
from random import random
x = [i for i in range(10)]
print(x)
y1 = [random() for i in range(10)]
y2 = [random() for i in range(10)]

plt.plot(x,y1,label="precision")
plt.plot(x,y2,label="recall")
plt.legend()
plt.show()
```
