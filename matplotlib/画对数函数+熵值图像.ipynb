```python
'''画出熵值的图像
'''
from matplotlib import pyplot as plt
import math
import numpy as np
x = np.arange(0.00001,1,0.000001) # 注意区间，因为logx中的x是
y = []
for i in x:    
    temp = -i*math.log(i) # 这个表达式就是熵值
    y.append(temp)
plt.plot(x,y)  # Plot some data on the axes.
plt.show()
```

```python
'''计算focal_loss 的表达式
01. 可以看到在(0,1)区间是一个单减函数
'''
from matplotlib import pyplot as plt
import math
import numpy as np
x = np.arange(0.001,1,0.001) # 注意区间，因为logx中的x是
y = []
gamma = 2
for i in x:
    temp = pow((1-i),gamma)
    temp = -temp * math.log(i) # $(1-p)^2 log p$
    y.append(temp)
plt.plot(x,y)  # Plot some data on the axes.
plt.show()
```
