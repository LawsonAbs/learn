```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker
from matplotlib import rc
    
'''可视化attention 的结果
一张图中画下所有子图
'''
def drawAttention():    
    fig = plt.figure(figsize = (8,10))# 画图，这里调整了图片的大小，否则会因为word太长导致文字重叠
    plt.rc('text', usetex=True)
    plt.rc('font', family='serif')
    
    # 画子图1
    df = pd.DataFrame([0.9432,0.8742,0.2739,0.0766,0.0576], 
                      columns=['lick'],
                      index=['*relief\%1:12:00::',
                             '*relief\%1:06:00::',
                             'relief\%1:28:00::',
                             'relief\%1:04:01::',
                             'relief\%1:04:02::'])    
        
    ax_1 = fig.add_subplot(2,2,1)
    cax = ax_1.matshow(df, interpolation='nearest', cmap='hot_r')
    fig.colorbar(cax)
    tick_spacing = 1
    ax_1.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
    ax_1.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
    ax_1.set_xticklabels([''] + list(df.columns),fontsize=14)
    ax_1.set_yticklabels([''] + list(df.index),fontsize=14)
    ax_1.set_title(r"When ancient wall sculptors \\were finished it was a \underline{relief}.",
                   fontsize=13,pad=30,verticalalignment='center',
                  horizontalalignment = 'center')
        
    # 画子图3
    df = pd.DataFrame([0.8434,0.0480,0.0011,0.0006,0.0006],
                      columns=['toast'],
                      index=['*frank\%5:00:00:\ndirect:02','*frank\%1:13:00::',
                             'frank\%1:18:00::','frank\%2:32:00::','frank\%5:00:00::\nobvious:00']
                     )
    ax_3 = fig.add_subplot(2,2,2)
    cax_3 = ax_3.matshow(df, interpolation='nearest', cmap='hot_r')
    fig.colorbar(cax_3)
    tick_spacing = 1
    ax_3.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
    ax_3.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
    ax_3.set_xticklabels([''] + list(df.columns),fontsize=14)
    ax_3.set_yticklabels([''] + list(df.index),fontsize=14)
    
    # pad的含义是标题和图像之间的距离
    ax_3.set_title(r"The  hot dog vendor answered,\\ with a tone quite \underline{frank}.",
                   fontsize=13,pad=30,
                  horizontalalignment = 'center',
                  verticalalignment='center')
    
    name='./attention.png'
    # 300参数用于保真
    plt.savefig(name,dpi=200 ) # 存储图片
    plt.show()
    
drawAttention()
```
