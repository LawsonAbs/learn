```python
"""读取xlsx文件
"""
import xlrd
import re # 实现对字符串按照多个字符分割

def readXlsx(filePath):
    s = xlrd.open_workbook(filePath)    
    sheet2 = s.sheet_by_index(1) # 获取第二张表格
    #print(sheet2)
    count_rows = sheet2.nrows # 获取第二张sheet的行数
    #print(count_rows)
#     count_ncols = sheet2.ncols # 获取列数
#     print(count_ncols)
    firstRow = sheet2.row_values(0)  # 获取第0行的内容    
    kwCnt ={} # 计算每个关键字对应的标签（是个列表的形式）
    for i in range(1,count_rows): # 遍历每行的内容        
        label = sheet2.cell(i,1) # 打印第2列的内容
        col3 = sheet2.cell(i,2) # 获取第3列的内容        
        col3 = str(col3)        
        col3 = col3.strip("text:'")                
        col3 = re.split('[|、]',col3)
        label = str(label).strip("text:'")
        for words in col3:
            # 计算每个关键词对应的标签数
            if words not in kwCnt.keys():
                kwCnt[words] = []
                kwCnt[words].append(label)
            else:
                if label not in kwCnt[words]: # 如果不在其中
                    kwCnt[words].append(label)
         #   print(words,kwCnt[words])
        #print() # 打出换行        
    #print(kwCnt)
    # 按照value进行排序    
    kwCnt = dict(sorted(kwCnt.items(),key = lambda kwCnt:len(kwCnt[1]),reverse=True))
    print("关键词","标签列表")
    for item in kwCnt.items():
        print(item[0],"=>",item[1])
    
filePath = "/Users/gamidev/program/financial/data/企业新闻舆情正负面标签_关键词参照.xlsx"
readXlsx(filePath)
        
```
